```python
url = 'http://dataservices.imf.org/REST/SDMX_JSON.svc/DataFlow'
series = requests.get(url)
imf_dfs = []
for d in series.json()['Structure']['Dataflows']['Dataflow']:
    imf_dfs.append([d['@id'].split('-')[1], d['Name']['#text']])
```

```python
def get_db_info(id):
    imf_dfs_codes = []
    url_d = 'http://dataservices.imf.org/REST/SDMX_JSON.svc/DataStructure/' + id
    try:
        series_d = requests.get(url_d)
        for c in series_d.json()['Structure']['CodeLists']['CodeList']:
            imf_dfs_codes.append([c['@id'], c['Name']['#text']])
    except:
        pass
    return imf_dfs_codes, series_d
```
