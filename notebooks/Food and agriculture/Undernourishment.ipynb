```python
# Prevalence of undernourishment (% od the population)
id_undern = 'SN.ITK.DEFC.ZS'
id_pop = 'SP.POP.TOTL'

def get_group_w(group):
    years = list(map(str, list(range(2017,2019))))
    groups = [[group]]

    data_df = get_wb_tb(years, id_undern, groups, export_to_excel=False)    
    weight_df = get_wb_tb(years, id_pop, groups, export_to_excel=False)
    dcs_w = wa(data_df, weight_df, 'entity' )
    
    return dcs_w, data_df, weight_df
```
