```python
from yahoofinancials import YahooFinancials
import pandas as pd
import json
import altair as alt
```

```python
def show_a_symbol(symbol):
    
    yahoo_financials = YahooFinancials(symbol)
    data = yahoo_financials.get_historical_price_data('2020-01-01', '2020-04-26', 'daily')
    data = pd.json_normalize(data['AAPL']['prices'])
    line1 = alt.Chart(data).mark_line().encode(x='yearmonthdate(formatted_date):T', y='close:Q').properties( width = 700)
    
    return line1
```

```python
show_a_symbol('AAPL')
```
