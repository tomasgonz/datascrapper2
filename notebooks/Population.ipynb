```python
def get_projection_by_groups(groups):
    datasets = []
    x = gpp(groups[0]).get_column_names()[2:len(gpp(groups[0]).get_column_names())]
    for g in groups:    
        y = gpp(g).get_total_by_column()[2:len(gpp(g).get_column_names())]
        datasets.append([x, y, g, 'Projected population, thousands'])
    
    return datasets
```
