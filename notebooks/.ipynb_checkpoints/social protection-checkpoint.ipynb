```python
"""
    Calculates weigthed average

    Parameters
    ----------
    data : Pandas dataframe
        Dataframe with data to be wieghted
    weight : Pandas dataframe
        Dataframe containing weight values
    data_field: string
        Name of field with data to weight
    weight_field: string
        Name of field with data for weighting
    name_weighted_column: string
        Name of column that will be created with weighted values
        
    Returns
    -------
    object
        Return a PANDAS dataframe with an additional column 
        containing weighted values

    """
def weighted_average(**kwargs):
    
    # Work with a copy of the original dataset so when we add
    # the column with the weights we do not get a warning
    data = kwargs['data'].copy()
    # Dataframe containing values for weigthing. It can be the same
    # than the dataframe with the data
    weight = kwargs['weight']
    # Sum all weight values
    sum_valid_weight = weight[kwargs['weight_field']].sum()
    # Calculate weigthed average
    c = data[kwargs['data_field']] * (weight[kwargs['weight_field']] / sum_valid_weight)
    # Add a column with weigthed values to the dataframe
    data[kwargs['name_weighted_column']] = c
    # Return dataframe with weighted values
    
    return data
```

```python
weighted = weighted_average(data=cov_ldcs_latest_2009, weight=cov_ldcs_latest_2009, data_field = "value", weight_field = "population", name_weighted_column="weight")
```
