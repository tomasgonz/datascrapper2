```python
import sys
sys.path.append("..")
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('../data/cces18_common_vv.csv')
```

```python
from scipy import stats
import sklearn
from sklearn.linear_model import LinearRegression
import statsmodels
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.formula.api import logit
from statsmodels.formula.api import mnlogit
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ind = 'CC18_320c Gun Control' # Ban assault rifles
model = str()"CC18_317 ~ " + ind)
voting_model = mnlogit(model , data = data).fit()
```

```python
voting_model.summary()
```

```python
fig = plt.figure(figsize=(15,8))
fig = sm.graphics.plot_regress_exog(voting_model, 'CC18_322d_new' , fig=fig)
```
