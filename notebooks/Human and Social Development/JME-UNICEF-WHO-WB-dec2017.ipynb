```python
import sys
sys.path.append("..")
import csv
import json
import urllib.request
from IPython.display import clear_output
import sources
from countrylist import CountryList
from data.Frame import Frame
from data.Row import Row
from data.Cell import Cell
from data.Stats import calculate_weighted_average as wa
c = CountryList()
c.load_wb()
from sources.worldbank.indicators import get_data_frame_wide as getdfw
```

```python
column_names = []
df = Frame()
with open('jme-dec-2017.csv') as csvfile:
    reader = csv.reader(csvfile, delimiter=',', quotechar='"')
    i = next(reader)
    for col in i:
        column_names.append(col)
    
    column_names[0] = 'iso3code'
    
    for r in reader:
        dr = Row()
        for i in range(0,len(r)):
            dr.cells.append(Cell(column=column_names[i], value=r[i]))
        
        df.rows.append(dr)
```

```python
groups=[['LDCs', 'Africa'],['LDCs', 'Asia and the Pacific'],['LDCs', 'America']]
```

```python
# We will store the final result here
ndf = Frame()
# Let's prepare the data to be properly formatted by region

for g in groups:              
    for ctry in c.get_groups(g):                
        years = []
        for r in df.search("iso3code", ctry.iso3code).rows:
            years.append(r.get_by_column_name('year').value.value)                               
        
        tdf = Frame()
        #tdf.rows = df.search("iso3code", ctry.iso3code).rows
        
        #ndf.rows.append(tdf.search("year", sorted(years)[len(years)-1]).rows[0])
        for r in df.search("iso3code", ctry.iso3code).rows:
            ndf.rows.append(r)
```
